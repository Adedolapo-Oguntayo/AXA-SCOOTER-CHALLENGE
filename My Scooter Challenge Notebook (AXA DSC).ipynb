```python
train.drop(['FirstName', 'LastName', 'City', 'StateProvinceName','PostalCode','PhoneNumber'], axis=1, inplace=True)
testdata.drop(['FirstName', 'LastName', 'City', 'StateProvinceName','PostalCode','PhoneNumber'], axis=1, inplace=True)
```

```python
train['Occupation'] = pd.Categorical(train['Occupation'])
train['Occupation'] = train['Occupation'].cat.codes


train['Education'] = pd.Categorical(train['Education'])
train['Education'] = train['Education'].cat.codes

train['CountryRegionName'] = pd.Categorical(train['CountryRegionName'])
train['CountryRegionName'] = train['CountryRegionName'].cat.codes
```

```python
testdata['Occupation'] = pd.Categorical(testdata['Occupation'])
testdata['Occupation'] = testdata['Occupation'].cat.codes


testdata['Education'] = pd.Categorical(testdata['Education'])
testdata['Education'] = testdata['Education'].cat.codes

testdata['CountryRegionName'] = pd.Categorical(testdata['CountryRegionName'])
testdata['CountryRegionName'] = testdata['CountryRegionName'].cat.codes
```

```python
testdata['Gender'] = testdata['Gender'].replace({'M':1, 'F':0})
testdata['MaritalStatus'] = testdata['MaritalStatus'].replace({'M':1, 'S':0})

train['Gender'] = train['Gender'].replace({'M':1, 'F':0})
train['MaritalStatus'] = train['MaritalStatus'].replace({'M':1, 'S':0})
```

```python
#Write a code to plot graph with dolaplot
def dolaplot(feature):
        phaset = sns.FacetGrid(train, hue = 'BuyScooter', aspect = 4)
        phaset.map(sns.kdeplot, feature, shade = True)
        phaset.set(xlim=(0, train[feature].max()))
        phaset.add_legend()
```

```python
dolaplot('YearlyIncome')
```

```python
def AgeDOLA(train):
    if train.YearlyIncome <=86000:
        return 'low_earners'
    else:
        return 'high_earners'
train['income_class'] = train.apply(lambda train:AgeDOLA(train), axis=1)
```

```python
def AgeDOLA(testdata):
    if testdata.YearlyIncome <=86000:
        return 'low_earners'
    else:
        return 'high_earners'
testdata['income_class'] = testdata.apply(lambda testdata:AgeDOLA(testdata), axis=1)
```

```python
train.head()
```

```python
#This is the code that transforms all the data from ranges to integer data type for binary only
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
train['income_class'] = le.fit_transform(train['income_class'])
```

```python
testdata['income_class'] = le.fit_transform(testdata['income_class'])
```

```python
#correlelogram- plot heatmap to find correlation among features (to identify what feature to drop and retain)
from matplotlib import pyplot as plt
corrmat =train.corr()
f, ax = plt.subplots(figsize=(20,10))
sns.heatmap(corrmat, square=True, annot=True,linewidth=0.8, cmap='RdBu')
```

```python
train.drop(['income_class','CountryRegionName'], axis=1, inplace=True)
testdata.drop(['income_class','CountryRegionName'], axis=1, inplace=True)
```
